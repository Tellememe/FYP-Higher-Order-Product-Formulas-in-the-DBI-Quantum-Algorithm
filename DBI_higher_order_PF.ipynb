```python
def is_sparse(matrix, threshold=0.5):
    """
    Check if a matrix is sparse based on a density threshold.

    Parameters:
        matrix (list of lists): The matrix to be checked.
        threshold (float): The density threshold to consider the matrix sparse. Default is 0.5.

    Returns:
        bool: True if the matrix is sparse, False otherwise.
    """
    num_elements = len(matrix) * len(matrix[0])
    num_nonzero_elements = sum(1 for row in matrix for element in row if element != 0)
    density = num_nonzero_elements / num_elements
    return density < threshold



print("Is sparse:", is_sparse(H))  # Output: True
print("Is sparse:", is_sparse(H))   # Output: False
```
